```r
library(tidyverse) 
library(tibbletime)
library(anomalize) 
library(lubridate) 
library(changepoint)
library(rio)
```

```r
url = 'https://github.com/akarlinsky/world_mortality/raw/main/world_mortality.csv'
world_mort = import(url)
world_mort
```

```r
rm = dplyr::filter(world_mort, country_name == 'Russia')
rm
```

```r
rm2 = mutate(rm, date = ymd(paste0(year, '-', time, '-01')))
```

```r
rm3 = dplyr::select(rm2, deaths, date)
```

```r
rm3
```

```r
qplot(data=rm3, x=date, y=deaths, geom='line')
```

```r
rm4 = as_tbl_time(rm3, index=date)
```

```r
one_break <- cpt.mean(rm4$deaths, method="AMOC")
```

```r
plot(one_break)
```

```r
all_break <- cpt.mean(rm4$deaths, method="BinSeg", Q=3)
```

```r
plot(all_break)
```

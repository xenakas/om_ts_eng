```r
library(fpp3) 
library(caret)
library(ranger)
library(xgboost)
```

```r
str(AirPassengers)
```

```r
air = as_tsibble(AirPassengers)
air
```

```r
gg_tsdisplay(air)
```

```r
air2 = mutate(air, 
    ln_pass = log(value),
    t = 1:nrow(air))
air2
```

```r
fourier_x = 
  forecast::fourier(AirPassengers, K = 2)
fourier_x
# sin(2pi t/12) cos(2pi t/12)  
# sin(2pi t/12 *2) cos(2pi t/12 *2)
```

```r
colnames(fourier_x) = c('s1', 'c1', 's2', 'c2')
```

```r
air3 = bind_cols(air2, fourier_x)
tail(air3)
```

```r
air3test = tail(air3, 24)
air3train = head(air3, -24)
```

```r
cv_params = trainControl(method = 'cv',
                         number = 5)
```

```r
ols = train(ln_pass ~ t + s1 + c1 + s2 + c2, 
            data = air3train,
            trControl = cv_params,
            method = 'lm')
rf = train(ln_pass ~ t + s1 + c1 + s2 + c2, 
            data = air3train,
            trControl = cv_params,
            method = 'ranger',
           num.trees = 1000)
gb = train(ln_pass ~ t + s1 + c1 + s2 + c2, 
           data = air3train,
           trControl = cv_params,
           method = 'xgbTree')
```

```r
ols_fcst = predict(ols, air3test)
rf_fcst = predict(rf, air3test)
gb_fcst = predict(gb, air3test)
```

```r
# air3train
```

```r
a <- air3test[['value']]-ols_fcst
b <- air3test[['value']]-rf_fcst
c <- air3test[['value']]-gb_fcst
```

```r
forecast::dm.test(a, b)
forecast::dm.test(b, c)
```
